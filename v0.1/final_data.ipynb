```python
import os
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
nim_jk = {}
nim_age = {}
data_raw = pd.read_csv("JenisKelaminSubjek.csv", delimiter=";", header=None)
# print(data_raw)

for i in range(len(data_raw)):
    nim_jk[data_raw[0][i]] = data_raw[2][i]
    nim_age[data_raw[0][i]] = data_raw[3][i]

print(nim_jk)
print(nim_age)
```

```python
data_final = pd.read_csv("psqi_and_backspan_result.csv", delimiter=",")
list_nim = list(data_final["NIM"])
print(list_nim)
sex = []
age = []
for student in list_nim:
    sex.append(nim_jk[student])
    age.append(nim_age[student])
print(sex)
print(age)
data_final["Sex"] = sex
data_final["Age"] = age
```
